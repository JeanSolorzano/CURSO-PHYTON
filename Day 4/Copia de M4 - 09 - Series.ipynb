# Ejercicio 4

Aplicar las operaciones aritméticas básicas sobre los objetos Series dados.

Prueba a:
* Sumar
* Restar
* Multiplicar
* Dividir

```python
 #"Este código crea dos series de números y realiza operaciones aritméticas básicas entre ellas, como suma, resta, multiplicación y división."
serie1 = pd.Series([1, 2, 3, 4, 5])   # Crea una serie llamada serie1 con los valores 1, 2, 3, 4 y 5
serie2 = pd.Series([9, 8, 7, 6, 5])    # Crea una serie llamada serie2 con los valores 9, 8, 7, 6 y 5

suma = serie1 + serie2  # Realiza la suma elemento a elemento de las dos series, creando una nueva serie llamada suma          
print(suma)     # Imprime la serie suma

resta = serie1 - serie2        # Realiza la resta elemento a elemento de las dos series, creando una nueva serie llamada resta       
print("Resta:", resta)        #imprime la serie de la resta

multiplicacion = serie1 * serie2            # Realiza la multiplicación elemento a elemento de las dos series, creando una nueva serie llamada multiplicacion
print("Multiplicación:", multiplicacion)     #imprime la serie de la multiplicacion

division = serie1 / serie2           # Realiza la multiplicación elemento a elemento de las dos series, creando una nueva serie llamada division
print("División:", division)         #imprime la serie de la divisioon

```

# Ejercicio 5

Usar operadores relacionales para comparar los objetos Series dados.



```python
serie1 = pd.Series([1, 2, 3, 5, 7])   # Crea una serie con los valores 1, 2, 3, 5 y 7
serie2 = pd.Series([13, 17, 11, 5, 3])   # Crea otra serie con los valores 13, 17, 11, 5 y 3

menor_que = serie1 < serie2          # Realiza una comparación elemento a elemento de ambas series, devolviendo una serie booleana con True donde el elemento de serie1 es menor que el correspondiente elemento de serie2
print("Menor que:", menor_que)    # Imprime la serie de booleanos

mayor_que = serie1 > serie2          # Realiza una comparación elemento a elemento de ambas series, devolviendo una serie booleana con True donde el elemento de serie1 es mayor que el correspondiente elemento de serie2
print("Mayor que:", mayor_que)         # Imprime la serie de booleanos

igual_a = serie1 == serie2           # Realiza una comparación elemento a elemento de ambas series, devolviendo una serie booleana con True donde el elemento de serie1 es igual al correspondiente elemento de serie2
print("Igual a:", igual_a)            # Imprime la serie de booleanos

distinto_de = serie1 != serie2         # Realiza una comparación elemento a elemento de ambas series, devolviendo una serie booleana con True donde el elemento de serie1 es diferente al correspondiente elemento de serie2
print("Distinto de:", distinto_de)     # Imprime la serie de booleanos
```
